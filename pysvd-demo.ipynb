```python
class Ui_PySVD(object):
    def setupUi(self, PySVD):
        PySVD.setObjectName("PySVD")
        PySVD.resize(800, 600)
        self.centralwidget = QtWidgets.QWidget(PySVD)
        self.centralwidget.setObjectName("centralwidget")
        self.category = QtWidgets.QLineEdit(self.centralwidget)
        self.category.setGeometry(QtCore.QRect(410, 290, 161, 31))
        self.category.setText("")
        self.category.setObjectName("category")
        self.TRAINbutton = QtWidgets.QPushButton(self.centralwidget)
        self.TRAINbutton.setGeometry(QtCore.QRect(520, 210, 131, 33))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(18)
        font.setBold(True)
        font.setWeight(75)
        self.TRAINbutton.setFont(font)
        self.TRAINbutton.setObjectName("TRAINbutton")
        self.label_fps = QtWidgets.QLabel(self.centralwidget)
        self.label_fps.setGeometry(QtCore.QRect(450, 150, 221, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(10)
        font.setBold(True)
        font.setWeight(75)
        self.label_fps.setFont(font)
        self.label_fps.setObjectName("label_fps")
        self.result = QtWidgets.QPlainTextEdit(self.centralwidget)
        self.result.setGeometry(QtCore.QRect(410, 450, 371, 71))
        self.result.setObjectName("result")
        self.label2_3 = QtWidgets.QLabel(self.centralwidget)
        self.label2_3.setGeometry(QtCore.QRect(410, 410, 91, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(12)
        font.setBold(False)
        font.setWeight(50)
        self.label2_3.setFont(font)
        self.label2_3.setObjectName("label2_3")
        self.frame = QtWidgets.QFrame(self.centralwidget)
        self.frame.setGeometry(QtCore.QRect(200, 30, 581, 51))
        self.frame.setFrameShape(QtWidgets.QFrame.StyledPanel)
        self.frame.setFrameShadow(QtWidgets.QFrame.Raised)
        self.frame.setObjectName("frame")
        self.label_train = QtWidgets.QLabel(self.frame)
        self.label_train.setGeometry(QtCore.QRect(10, 10, 161, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(10)
        font.setBold(True)
        font.setWeight(75)
        self.label_train.setFont(font)
        self.label_train.setObjectName("label_train")
        self.trainfolder = QtWidgets.QLineEdit(self.frame)
        self.trainfolder.setGeometry(QtCore.QRect(160, 10, 311, 31))
        self.trainfolder.setObjectName("trainfolder")
        self.browse_train = QtWidgets.QPushButton(self.frame)
        self.browse_train.setGeometry(QtCore.QRect(480, 10, 93, 31))
        self.browse_train.setObjectName("browse_train")
        self.frame_2 = QtWidgets.QFrame(self.centralwidget)
        self.frame_2.setGeometry(QtCore.QRect(200, 90, 581, 51))
        self.frame_2.setFrameShape(QtWidgets.QFrame.StyledPanel)
        self.frame_2.setFrameShadow(QtWidgets.QFrame.Raised)
        self.frame_2.setObjectName("frame_2")
        self.label_test = QtWidgets.QLabel(self.frame_2)
        self.label_test.setGeometry(QtCore.QRect(10, 10, 161, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(10)
        font.setBold(True)
        font.setWeight(75)
        self.label_test.setFont(font)
        self.label_test.setObjectName("label_test")
        self.testfolder = QtWidgets.QLineEdit(self.frame_2)
        self.testfolder.setGeometry(QtCore.QRect(160, 10, 311, 31))
        self.testfolder.setObjectName("testfolder")
        self.browse_test = QtWidgets.QPushButton(self.frame_2)
        self.browse_test.setGeometry(QtCore.QRect(480, 10, 93, 31))
        self.browse_test.setObjectName("browse_test")
        self.log = QtWidgets.QTextBrowser(self.centralwidget)
        self.log.setGeometry(QtCore.QRect(20, 150, 381, 241))
        self.log.setObjectName("log")
        self.logo = QtWidgets.QLabel(self.centralwidget)
        self.logo.setGeometry(QtCore.QRect(10, 20, 181, 91))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(26)
        font.setBold(True)
        font.setWeight(75)
        self.logo.setFont(font)
        self.logo.setObjectName("logo")
        self.label_snapshots = QtWidgets.QLabel(self.centralwidget)
        self.label_snapshots.setGeometry(QtCore.QRect(410, 260, 181, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(10)
        self.label_snapshots.setFont(font)
        self.label_snapshots.setObjectName("label_snapshots")
        self.label_category = QtWidgets.QLabel(self.centralwidget)
        self.label_category.setGeometry(QtCore.QRect(590, 260, 121, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(10)
        self.label_category.setFont(font)
        self.label_category.setObjectName("label_category")
        self.pushButton = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton.setGeometry(QtCore.QRect(720, 290, 71, 31))
        self.pushButton.setObjectName("pushButton")
        self.label = QtWidgets.QLabel(self.centralwidget)
        self.label.setGeometry(QtCore.QRect(20, 400, 381, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(12)
        self.label.setFont(font)
        self.label.setObjectName("label")
        self.label_2 = QtWidgets.QLabel(self.centralwidget)
        self.label_2.setGeometry(QtCore.QRect(20, 430, 351, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(12)
        self.label_2.setFont(font)
        self.label_2.setObjectName("label_2")
        self.label_3 = QtWidgets.QLabel(self.centralwidget)
        self.label_3.setGeometry(QtCore.QRect(110, 460, 331, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(12)
        self.label_3.setFont(font)
        self.label_3.setObjectName("label_3")
        self.label_4 = QtWidgets.QLabel(self.centralwidget)
        self.label_4.setGeometry(QtCore.QRect(170, 490, 291, 31))
        font = QtGui.QFont()
        font.setFamily("Georgia")
        font.setPointSize(12)
        self.label_4.setFont(font)
        self.label_4.setObjectName("label_4")
        self.category_2 = QtWidgets.QLineEdit(self.centralwidget)
        self.category_2.setGeometry(QtCore.QRect(410, 330, 161, 31))
        self.category_2.setText("")
        self.category_2.setObjectName("category_2")
        self.category_3 = QtWidgets.QLineEdit(self.centralwidget)
        self.category_3.setGeometry(QtCore.QRect(410, 370, 161, 31))
        self.category_3.setText("")
        self.category_3.setObjectName("category_3")
        self.pushButton_2 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_2.setGeometry(QtCore.QRect(720, 330, 71, 31))
        self.pushButton_2.setObjectName("pushButton_2")
        self.pushButton_3 = QtWidgets.QPushButton(self.centralwidget)
        self.pushButton_3.setGeometry(QtCore.QRect(720, 370, 71, 31))
        self.pushButton_3.setObjectName("pushButton_3")
        self.testfolder_2 = QtWidgets.QLineEdit(self.centralwidget)
        self.testfolder_2.setGeometry(QtCore.QRect(670, 150, 71, 31))
        self.testfolder_2.setObjectName("testfolder_2")
        self.testfolder_3 = QtWidgets.QLineEdit(self.centralwidget)
        self.testfolder_3.setGeometry(QtCore.QRect(590, 290, 111, 31))
        self.testfolder_3.setObjectName("testfolder_3")
        self.testfolder_4 = QtWidgets.QLineEdit(self.centralwidget)
        self.testfolder_4.setGeometry(QtCore.QRect(590, 330, 111, 31))
        self.testfolder_4.setObjectName("testfolder_4")
        self.testfolder_5 = QtWidgets.QLineEdit(self.centralwidget)
        self.testfolder_5.setGeometry(QtCore.QRect(590, 370, 111, 31))
        self.testfolder_5.setObjectName("testfolder_5")
        PySVD.setCentralWidget(self.centralwidget)
        self.menubar = QtWidgets.QMenuBar(PySVD)
        self.menubar.setGeometry(QtCore.QRect(0, 0, 800, 26))
        self.menubar.setObjectName("menubar")
        PySVD.setMenuBar(self.menubar)
        self.statusbar = QtWidgets.QStatusBar(PySVD)
        self.statusbar.setObjectName("statusbar")
        PySVD.setStatusBar(self.statusbar)

        self.retranslateUi(PySVD)
        self.TRAINbutton.clicked.connect(self.get_data)
        QtCore.QMetaObject.connectSlotsByName(PySVD)
        
    def get_data(self):
        from rom import ROM
        from rom import ML
        import numpy as np
        dir_train = self.trainfolder.text()
        dir_test = self.testfolder.text()
        freq = int(self.testfolder_2.text())
        category1_size = int(self.category.text())
        category2_size = int(self.category_2.text())
        category3_size = int(self.category_3.text())
        category1_name = self.testfolder_3.text()
        category2_name = self.testfolder_4.text()
        category3_name = self.testfolder_5.text()
        group = [category1_size,category2_size,category3_size]
        group_name = [category1_name,category2_name,category3_name]
        self.log.append('Reduced Ordered Model Created ...')
        self.log.append('Model Training Finished')
        dataset1 = ROM(dir_train,dir_test,freq)
        Up, Sp, Vp = dataset1.pca()
        U,S,V,X = dataset1.pod()
        C = dataset1.dft(V[:,[0]])
        NFFT = dataset1.nextPowerOf2()
        f, P1, max_freq = dataset1.PSD(C,NFFT)
        rank = dataset1.rank_trunc(S)
        Ur, Vr, Sr, Phi, D, omega, X_dmd, b, q,ind, t_dyn = dataset1.DMD(rank)
        error, error1 = dataset1.Error(X_dmd)
        coeff_PCA, coeff_POD, coeff_DMD, coeff_PCA_test, coeff_POD_test, coeff_DMD_test = dataset1.ROM_projected(Up,U,Phi,rank)
        np.savetxt('PCA_data.csv', coeff_PCA, delimiter=',')
        np.savetxt('POD_data.csv', coeff_POD, delimiter=',')
        np.savetxt('DMD_data.csv', coeff_DMD, delimiter=',')
        np.savetxt('PCA_data_test.csv', coeff_PCA_test, delimiter=',')
        np.savetxt('POD_data_test.csv', coeff_POD_test, delimiter=',')
        np.savetxt('DMD_data_test.csv', coeff_DMD_test, delimiter=',')
        directory1 = 'PCA_data.csv'
        directory2 = 'POD_data.csv'
        directory3 = 'DMD_data.csv'
        directory4 = 'PCA_data_test.csv'
        directory5 = 'POD_data_test.csv'
        directory6 = 'DMD_data_test.csv'
        
        dataset2 = ML(directory1,directory4,rank,group,group_name)
        y_pred_pca_svm, cm1_pca_svm, accuracy_train_pca_svm, y_pred_test_pca_svm, string_pca_svm, group_pca_svm, group_name_pca_svm, index_pca_svm, predict_pca_svm, y, X = dataset2.SVM()
        y_pred_pca_rf, cm1_pca_rf, accuracy_train_pca_rf, y_pred_test_pca_rf, string_pca_rf, group_pca_rf, group_name_pca_rf, index_pca_rf, predict_pca_rf = dataset2.RF()
        
        dataset3 = ML(directory2,directory5,rank,group,group_name)
        y_pred_pod_svm, cm1_pod_svm, accuracy_train_pod_svm, y_pred_test_pod_svm, string_pod_svm, group_pod_svm, group_name_pod_svm, index_pod_svm, predict_pod_svm, y, X = dataset3.SVM()
        y_pred_pod_rf, cm1_pod_rf, accuracy_train_pod_rf, y_pred_test_pod_rf, string_pod_rf, group_pod_rf, group_name_pod_rf, index_pod_rf, predict_pod_rf = dataset3.RF()
        
        dataset4 = ML(directory3,directory6,rank,group,group_name)
        y_pred_dmd_svm, cm1_dmd_svm, accuracy_train_dmd_svm, y_pred_test_dmd_svm, string_dmd_svm, group_dmd_svm, group_name_dmd_svm, index_dmd_svm, predict_dmd_svm, y, X = dataset4.SVM()
        y_pred_dmd_rf, cm1_dmd_rf, accuracy_train_dmd_rf, y_pred_test_dmd_rf, string_dmd_rf, group_dmd_rf, group_name_dmd_rf, index_dmd_rf, predict_dmd_rf  = dataset4.RF()
        
        strings = [string_pca_rf,string_dmd_svm,string_dmd_rf]
        result = [0]*len(group_name)
        for i in range(0,len(group_name)):
            result[i] = strings.count(group_name[i])
            string = np.argmax(result)
            final = group_name[string]
        self.result.appendPlainText('This dataset belongs to the'+' '+final +' '+'category') # +' '+'with a dominant maximum frequency of'+' '+str(max_freq))

    def retranslateUi(self, PySVD):
        _translate = QtCore.QCoreApplication.translate
        PySVD.setWindowTitle(_translate("PySVD", "MainWindow"))
        self.TRAINbutton.setText(_translate("PySVD", "START "))
        self.label_fps.setText(_translate("PySVD", "Set Frames per Second"))
        self.label2_3.setText(_translate("PySVD", "RESULTS"))
        self.label_train.setText(_translate("PySVD", "TRAIN FOLDER"))
        self.browse_train.setText(_translate("PySVD", "BROWSE"))
        self.label_test.setText(_translate("PySVD", "TEST FOLDER"))
        self.browse_test.setText(_translate("PySVD", "BROWSE"))
        self.logo.setText(_translate("PySVD", "PySVD "))
        self.label_snapshots.setText(_translate("PySVD", "Number of snapshots "))
        self.label_category.setText(_translate("PySVD", "Category Name"))
        self.pushButton.setText(_translate("PySVD", "ADD"))
        self.label.setText(_translate("PySVD", "Created By:           Keivan Mokhtarpour "))
        self.label_2.setText(_translate("PySVD", "Contact:           m.keivan73@gmail.com"))
        self.label_3.setText(_translate("PySVD", "          ali.dolatabadi@concordia.ca"))
        self.label_4.setText(_translate("PySVD", "        jadidi_m@yahoo.com"))
        self.pushButton_2.setText(_translate("PySVD", "ADD"))
        self.pushButton_3.setText(_translate("PySVD", "ADD")) 
```
